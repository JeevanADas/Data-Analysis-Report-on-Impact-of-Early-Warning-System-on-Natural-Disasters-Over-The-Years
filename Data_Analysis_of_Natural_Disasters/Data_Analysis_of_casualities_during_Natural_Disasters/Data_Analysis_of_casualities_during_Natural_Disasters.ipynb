# Reading data from the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("deaths-natural-disasters.csv")
df
```

# Deleting Columns that are not needed for our study


```python
del df['Country Name']
del df['Country Code']
df.head()
```

# Making a new Data frame with the number of casualites every year 
### (1900 - 2013)

```python
addeddead = df.sum(axis = 0, skipna = True)
dead = pd.DataFrame(addeddead, columns = ['Casualties']).sort_values(by='Casualties', ascending=False)
```

```python
dead
```

```python
dead = dead.head(10)
dead.plot(y='Casualties', kind='bar', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Casualties in millions------")
plt.title("\n Figure 1.1 \n Top 10 years with the most number of casualties (1900 - 2013)")
```

# Making a new Data frame with the amount of people injured in the recent years
### (2013 - 2004)

```python
dead_desc = pd.DataFrame(addeddead, columns = ['Casualties'])
dead_desc = dead_desc.head(10)
dead_desc
```

```python
dead_desc.plot(y='Casualties', kind='line', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Casualties in millions------")
plt.title("\n Figure 1.2 \n Casualties in the recent years (2013 - 2004)")
```

# Inference

In figure 1.1, we can observe that all the years present in top 10 are amongst the 1900s, when the early warning facities were under-developed as compared to the recent years, i.e 2000s as shown in Figure 1.2.
