# Reading data from the datasheet

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("homeless-natural-disasters.csv")
df
```

# Deleting Columns that are not needed for our study

```python
del df['Country Name']
del df['Country Code']
df.head()
```

# Making a new dataframe with the number of people left homeless every year
### (1900 - 2013)

```python
addedhomeless = df.sum(axis = 0, skipna = True)
homeless = pd.DataFrame(addedhomeless, columns = ['Homeless']).sort_values(by='Homeless', ascending=False)
```

```python
homeless
```

```python
homeless = homeless.head(10)
homeless.plot(y='Homeless', kind='bar', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Homeless in millions------")
plt.title("\n Figure 4.1 \n Top 10 years with the most number of people left Homeless (1900 - 2013)")
```

# Making a new Data frame with the amount of people injured in the recent years
### (2013 - 2004)

```python
homeless_desc = pd.DataFrame(addedhomeless, columns = ['Homeless'])
homeless_desc = homeless_desc.head(10)
homeless_desc
```

```python
homeless_desc.plot(y='Homeless', kind='line', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Homeless in millions------")
plt.title("\n Figure 4.2 \n  Number of people left Homeless due to natural disasters in the recent years (2013 - 2004)")
```

# Inference

As seen in figure 4.1, the top 10 years are amongst the 1900s. One main reason of that are the poor infrastructure facitilies as compared to the mordern times. The infrastructure failed to withhold itself in the course of a natural disaster, leaving people without a home.

Figure 4.2 indicates the lessening of homeless people due to natural disasters around the world in the modern years with better infrastructure facilities.
