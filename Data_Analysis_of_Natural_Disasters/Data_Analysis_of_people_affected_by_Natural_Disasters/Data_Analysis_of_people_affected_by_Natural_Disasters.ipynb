# Reading data from the datasheet

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("total-affected-natural-disasters.csv")
df
```

# Deleting Columns that are not needed for our study

```python
del df['Country Name']
del df['Country Code']
df.head()
```

# Making a new dataframe with the number of people affected by natural disaster every year

```python
addedaffected = df.sum(axis = 0, skipna = True)
affected = pd.DataFrame(addedaffected, columns = ['Affected']).sort_values(by='Affected', ascending=False)
```

```python
affected
```

```python
affected = affected.head(10)
affected.plot(y='Affected', kind='bar', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Affected in millions------")
plt.title("\n Figure 5.1 \n Top 10 years with the most number of people Affected (1900 - 2013)")
```

# Making a new Data frame with the amount of people Affected in the recent years
### (2013 - 2004)

```python
affected_desc = pd.DataFrame(addedaffected, columns = ['Affected'])
affected_desc = affected_desc.head(10)
affected_desc
```

```python
affected_desc.plot(y='Affected', kind='line', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Affected in millions------")
plt.title("\n Figure 5.2 \n  Number of people affected by natural disasters in the recent years (2013 - 2004)")
```

# Inference

In figure 5.1, 2002 is the highest in the list due to record-breaking rains, triggering devastating floods in Europe, destruction of homes across the Caribbean and life-threatening mudslides in India, Nepal and Bangladesh.  

Similarly, as mentioned in the analysis of the number of people injured due to natural disasters, the total amount of people affected in 2010 is high due to the earthquake in Haiti and the Russian summer heat wave.  

Following the year 2010, the graph declines wrt the amount of people being affected due to the development in early warning technologies.
