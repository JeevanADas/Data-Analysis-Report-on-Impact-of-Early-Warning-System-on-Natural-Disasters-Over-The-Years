# Reading data from the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("injured-natural-disasters.csv")
df
```

# Deleting Columns that are not neccessary for our study

```python
del df['Country Name']
del df['Country Code']
df.head()
```

# Making a new Data frame with the number of people injured every year 
### (1900 - 2013)

```python
addedinjured = df.sum(axis = 0, skipna = True)
injured = pd.DataFrame(addedinjured, columns = ['Injured']).sort_values(by='Injured', ascending=False)
```

```python
injured
```

```python
injured = injured.head(10)
injured.plot(y='Injured', kind='bar', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Injured in millions ------")
plt.title("\n Figure 2.1 \n Top 10 years with the most number of people injured(1900 - 2013)")
```

# Making a new Data frame with the amount of people injured in the recent years
### (2013 - 2004)

```python
injured_desc = pd.DataFrame(addedinjured, columns = ['Injured'])
injured_desc = injured_desc.head(10)
injured_desc
```

```python
injured_desc.plot(y='Injured', kind='line', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Injured in millions ------")
plt.title("\n Figure 2.2 \n The number of people injured in recent years (2013 - 2004)")
```

# Inference

In figure 2.1, we observe that 2004 is the highest year with the most number of injuries, this is because of the Indian Ocean earthquake and tsunami, making it one of the deadliest years in modern history.
 
Similarly, the year 2010, despite of it being a recent year with more number of technologies that gave early warnings, the earthquake in Haiti and the Russian summer heat wave caused multiple injuries and fatalities, giving it a high rank in the list.
