# Reading data from the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("total-damage-natural-disasters.csv")
df
```

# Deleting Columns that are not neccessary for our study

```python
del df['Country Name']
del df['Country Code']
df.head()
```

# Making a new Data frame with aggregated sum of damage caused by natural disasters every year 
### (1900 - 2013)

```python
damage = df.sum(axis = 0, skipna = True)
damage_asc = pd.DataFrame(damage, columns = ['Damage Caused']).sort_values(by='Damage Caused', ascending=False)
```

```python
damage_asc
```

```python
damage_asc = damage_asc.head(10)
damage_asc.plot(y='Damage Caused', kind='bar', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Damage caused by natural disaster in USD ------")
plt.title("\n Figure 3.1 \n Top 10 years with the most damage caused by natural disasters(1900 - 2013)")
```

# Making a new Data frame with aggregated sum of damage caused by natural disasters in the recent years
### (2013 - 2004)

```python
damage_desc = pd.DataFrame(damage, columns = ['Damage Caused'])
damage_desc = damage_desc.head(10)
damage_desc
```

```python
damage_desc.plot(y='Damage Caused', kind='line', figsize = (10,5))
plt.xlabel("-------- Year --------")
plt.ylabel("------ Damage caused by natural disaster in USD ------")
plt.title("\n Figure 3.2 \n Damage caused by natural disasters in the recent years(2013 - 2004)")
```

# Inference

In figure 3.1, we observe that most of the damage costs are high in the 2000s due to more developed and mordern infrastucture which increases the total cost as compared to the 1900s. Another factor that effects the damage cost is the increasing population in the cities over the years.
